# Python Insights - Analisando Dados com Python

### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
import pandas as pd
import plotly.express as px

# Passo a passo do projeto
# Passo 1: Importar a base de dados
# Passo 2: Visualizar a base de dados
# Passo 3: Tratamento de errod (resolver os erros na base)
# Passo 3.1 tratar valores vazios
# Passo 3.2 deletar informações insignificantes para analise: nome da inha/coluna entre "" + axis=0 deleta linha axis=1 deleta coluna
# Passo 4: Analise inicial dos dados ( entender o cancelamento)
# Passo 5: Analise profunda da base de dados ( encontrar a causa do cancelamento)

```

```python
# Passo 1: Importar a base de dados
tabela=pd.read_csv("cancelamentos.csv")
```

```python
# Passo 2: Visualizar a base de dados
display(tabela)
```

```python
# Passo 3: Tratamento de erros (resolver os erros na base)
display(tabela.info())
tabela=tabela.dropna()
tabela=tabela.drop("CustomerID", axis=1)
```

```python
# Passo 4: Analise inicial dos dados ( entender o cancelamento)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# analisando o contrato mensal
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
# descobrimos aqui que a média de cancelamentos é 1, ou seja, praticamente todos os contratos mensais cancelaram (ou todos)
# Constatado que contrato mensal sempre cancela, trabalhar no porqu
```

```python
# Passo 5: Analise profunda da base de dados ( encontrar a causa do cancelamento)

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
```
